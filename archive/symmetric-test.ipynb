```python
import os
import time
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives import padding
import secrets
import warnings
import pandas as pd
from scipy.stats import f_oneway
from tabulate import tabulate
import matplotlib.pyplot as plt


def generate_key_and_iv(algorithm):
    """Generate an appropriate key and IV based on the algorithm."""
    if algorithm in (algorithms.AES, algorithms.ChaCha20):
        key = secrets.token_bytes(32)  # 256-bit key
    elif algorithm == algorithms.TripleDES:
        key = secrets.token_bytes(24)  # 192-bit key (for 3DES)
    elif algorithm == algorithms.Blowfish:
        key = secrets.token_bytes(16)  # 128-bit key (max for Blowfish)
    else:
        raise ValueError("Unsupported algorithm")

    if algorithm == algorithms.Blowfish:
        iv = secrets.token_bytes(8)  # 64-bit IV for Blowfish
    elif algorithm in (algorithms.AES, algorithms.TripleDES):
        iv = secrets.token_bytes(16)  # 128-bit IV for block ciphers
    else:
        iv = secrets.token_bytes(16)  # 128-bit nonce for ChaCha20

    return key, iv


def encrypt_file(file_path, algorithm):
    """Encrypt the file using the given algorithm and measure the time."""
    with open(file_path, "rb") as f:
        data = f.read()

    key, iv = generate_key_and_iv(algorithm)
    if algorithm in (algorithms.AES, algorithms.TripleDES, algorithms.Blowfish):
        cipher = Cipher(algorithm(key), modes.CBC(iv), backend=default_backend())
        padder = padding.PKCS7(algorithm.block_size).padder()
        padded_data = padder.update(data) + padder.finalize()
        encryptor = cipher.encryptor()

        start_time = time.time()
        encrypted_data = encryptor.update(padded_data) + encryptor.finalize()
        end_time = time.time()
    elif algorithm == algorithms.ChaCha20:
        cipher = Cipher(algorithm(key, iv), mode=None, backend=default_backend())
        encryptor = cipher.encryptor()

        start_time = time.time()
        encrypted_data = encryptor.update(data)
        end_time = time.time()

    encryption_time = end_time - start_time
    return encryption_time


# Define file paths for text, video, and audio files
sample_text_files = [
    f"input_data/sample_text/text_{size}.txt"
    for size in [10, 100, 1000, 10000, 100000, 1000000, 10000000]
]

video_resolutions = ['1280x720']
video_sizes = [1, 2, 5, 10, 20, 30]
sample_video_files = [
    f"input_data/sample_video/SampleVideo_{res}_{size}mb.mp4"
    for res in video_resolutions
    for size in video_sizes
]

sample_audio_files = [
    "input_data/sample_audio/file_example_MP3_1MG.mp3",
    "input_data/sample_audio/file_example_MP3_2MG.mp3",
    "input_data/sample_audio/file_example_MP3_5MG.mp3",
    "input_data/sample_audio/file_example_MP3_700KB.mp3"
]

# Algorithms to test
algorithms_to_test = {
    "AES": algorithms.AES,
    "3DES": algorithms.TripleDES,
    "Blowfish": algorithms.Blowfish,
    "ChaCha20": algorithms.ChaCha20,
}

# Encryption results
results_text = []
results_video = []
results_audio = []

# Process each file type
file_groups = {
    "Text": sample_text_files,
    "Video": sample_video_files,
    "Audio": sample_audio_files,
}

for category, file_list in file_groups.items():
    for file_path in file_list:
        if not os.path.exists(file_path):
            print(f"File not found: {file_path}")
            continue

        file_results = {"File": file_path}
        for algo_name, algo in algorithms_to_test.items():
            try:
                if algo_name == "Blowfish":
                    with warnings.catch_warnings():
                        warnings.simplefilter("ignore")  # Suppress Blowfish deprecation warnings
                        print(f"Encrypting {file_path} using {algo_name}...")
                        encryption_time = encrypt_file(file_path, algo)
                else:
                    print(f"Encrypting {file_path} using {algo_name}...")
                    encryption_time = encrypt_file(file_path, algo)
                file_results[algo_name] = encryption_time
            except Exception as e:
                print(f"Error encrypting {file_path} using {algo_name}: {e}")
                file_results[algo_name] = None

        if category == "Text":
            results_text.append(file_results)
        elif category == "Video":
            results_video.append(file_results)
        elif category == "Audio":
            results_audio.append(file_results)

# Convert results to DataFrames
df_text = pd.DataFrame(results_text)
df_video = pd.DataFrame(results_video)
df_audio = pd.DataFrame(results_audio)

# Display results using tabulate
print("\nText Encryption Time Comparison:")
print(tabulate(df_text, headers="keys", tablefmt="grid"))

print("\nVideo Encryption Time Comparison:")
print(tabulate(df_video, headers="keys", tablefmt="grid"))

print("\nAudio Encryption Time Comparison:")
print(tabulate(df_audio, headers="keys", tablefmt="grid"))

# ANOVA Analysis for each category
print("\nANOVA Analysis for Text Files:")
anova_data_text = {algo: df_text[algo].dropna().values for algo in algorithms_to_test.keys()}
anova_result_text = f_oneway(*anova_data_text.values())
print(f"F-statistic: {anova_result_text.statistic:.6f}, p-value: {anova_result_text.pvalue:.6f}")

print("\nANOVA Analysis for Video Files:")
anova_data_video = {algo: df_video[algo].dropna().values for algo in algorithms_to_test.keys()}
anova_result_video = f_oneway(*anova_data_video.values())
print(f"F-statistic: {anova_result_video.statistic:.6f}, p-value: {anova_result_video.pvalue:.6f}")

print("\nANOVA Analysis for Audio Files:")
anova_data_audio = {algo: df_audio[algo].dropna().values for algo in algorithms_to_test.keys()}
anova_result_audio = f_oneway(*anova_data_audio.values())
print(f"F-statistic: {anova_result_audio.statistic:.6f}, p-value: {anova_result_audio.pvalue:.6f}")

# Visualization
for df, category in [(df_text, "Text"), (df_video, "Video"), (df_audio, "Audio")]:
    df_long = df.melt(id_vars=["File"], var_name="Algorithm", value_name="Time")
    df_long = df_long.dropna()  # Remove rows with missing values

    plt.figure(figsize=(10, 6))
    for algo in algorithms_to_test.keys():
        algo_data = df_long[df_long["Algorithm"] == algo]
        plt.plot(algo_data["File"], algo_data["Time"], marker="o", label=algo)

    plt.title(f"{category} Encryption Time Comparison")
    plt.xlabel("Files")
    plt.ylabel("Encryption Time (s)")
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import os
import time
import secrets
import warnings
import pandas as pd
from scipy.stats import f_oneway
from tabulate import tabulate
import matplotlib.pyplot as plt
import crypto
from crypto.Cipher import AES, DES3, Blowfish, ChaCha20
from crypto.Random import get_random_bytes
from crypto.Util.Padding import pad, unpad


def generate_key_and_iv(algorithm):
    """Generate an appropriate key and IV/nonce based on the algorithm."""
    if algorithm == AES:
        key = get_random_bytes(32)  # 256-bit key
        iv = get_random_bytes(16)  # 128-bit IV
    elif algorithm == DES3:
        key = DES3.adjust_key_parity(get_random_bytes(24))  # 192-bit key
        iv = get_random_bytes(8)  # 64-bit IV for 3DES
    elif algorithm == Blowfish:
        key = get_random_bytes(16)  # 128-bit key
        iv = get_random_bytes(8)  # 64-bit IV for Blowfish
    elif algorithm == ChaCha20:
        key = get_random_bytes(32)  # 256-bit key
        nonce = get_random_bytes(12)  # 96-bit nonce
        return key, nonce  # ChaCha20 uses a nonce instead of an IV
    else:
        raise ValueError("Unsupported algorithm")
    return key, iv


def encrypt_file(file_path, algorithm):
    """Encrypt the file using the given algorithm and measure the time."""
    with open(file_path, "rb") as f:
        data = f.read()

    key, iv_or_nonce = generate_key_and_iv(algorithm)

    if algorithm == AES:
        cipher = AES.new(key, AES.MODE_CBC, iv_or_nonce)
        block_size = cipher.block_size
        padded_data = pad(data, block_size)
    elif algorithm == DES3:
        cipher = DES3.new(key, DES3.MODE_CBC, iv_or_nonce)
        block_size = cipher.block_size
        padded_data = pad(data, block_size)
    elif algorithm == Blowfish:
        cipher = Blowfish.new(key, Blowfish.MODE_CBC, iv_or_nonce)
        block_size = cipher.block_size
        padded_data = pad(data, block_size)
    elif algorithm == ChaCha20:
        cipher = ChaCha20.new(key=key, nonce=iv_or_nonce)
        padded_data = data  # No padding needed for ChaCha20
    else:
        raise ValueError("Unsupported algorithm")

    start_time = time.time()
    encrypted_data = cipher.encrypt(padded_data)
    end_time = time.time()

    encryption_time = end_time - start_time
    return encryption_time


# Define file paths for text, video, and audio files
sample_text_files = [
    f"input_data/sample_text/text_{size}.txt"
    for size in [10, 100, 1000, 10000, 100000, 1000000, 10000000]
]

video_resolutions = ['1280x720']
video_sizes = [1, 2, 5, 10, 20, 30]
sample_video_files = [
    f"input_data/sample_video/SampleVideo_{res}_{size}mb.mp4"
    for res in video_resolutions
    for size in video_sizes
]

sample_audio_files = [
    "input_data/sample_audio/file_example_MP3_1MG.mp3",
    "input_data/sample_audio/file_example_MP3_2MG.mp3",
    "input_data/sample_audio/file_example_MP3_5MG.mp3",
    "input_data/sample_audio/file_example_MP3_700KB.mp3"
]

# Algorithms to test
algorithms_to_test = {
    "AES": AES,
    "3DES": DES3,
    "Blowfish": Blowfish,
    "ChaCha20": ChaCha20,
}

# Encryption results
results_text = []
results_video = []
results_audio = []

# Process each file type
file_groups = {
    "Text": sample_text_files,
    "Video": sample_video_files,
    "Audio": sample_audio_files,
}

for category, file_list in file_groups.items():
    for file_path in file_list:
        if not os.path.exists(file_path):
            print(f"File not found: {file_path}")
            continue

        file_results = {"File": file_path}
        for algo_name, algo in algorithms_to_test.items():
            try:
                print(f"Encrypting {file_path} using {algo_name}...")
                encryption_time = encrypt_file(file_path, algo)
                file_results[algo_name] = encryption_time
            except Exception as e:
                print(f"Error encrypting {file_path} using {algo_name}: {e}")
                file_results[algo_name] = None

        if category == "Text":
            results_text.append(file_results)
        elif category == "Video":
            results_video.append(file_results)
        elif category == "Audio":
            results_audio.append(file_results)

# Convert results to DataFrames
df_text = pd.DataFrame(results_text)
df_video = pd.DataFrame(results_video)
df_audio = pd.DataFrame(results_audio)

# Display results using tabulate
print("\nText Encryption Time Comparison:")
print(tabulate(df_text, headers="keys", tablefmt="grid"))

print("\nVideo Encryption Time Comparison:")
print(tabulate(df_video, headers="keys", tablefmt="grid"))

print("\nAudio Encryption Time Comparison:")
print(tabulate(df_audio, headers="keys", tablefmt="grid"))

# ANOVA Analysis for each category
print("\nANOVA Analysis for Text Files:")
anova_data_text = {algo: df_text[algo].dropna().values for algo in algorithms_to_test.keys()}
anova_result_text = f_oneway(*anova_data_text.values())
print(f"F-statistic: {anova_result_text.statistic:.6f}, p-value: {anova_result_text.pvalue:.6f}")

print("\nANOVA Analysis for Video Files:")
anova_data_video = {algo: df_video[algo].dropna().values for algo in algorithms_to_test.keys()}
anova_result_video = f_oneway(*anova_data_video.values())
print(f"F-statistic: {anova_result_video.statistic:.6f}, p-value: {anova_result_video.pvalue:.6f}")

print("\nANOVA Analysis for Audio Files:")
anova_data_audio = {algo: df_audio[algo].dropna().values for algo in algorithms_to_test.keys()}
anova_result_audio = f_oneway(*anova_data_audio.values())
print(f"F-statistic: {anova_result_audio.statistic:.6f}, p-value: {anova_result_audio.pvalue:.6f}")

# Visualization
for df, category in [(df_text, "Text"), (df_video, "Video"), (df_audio, "Audio")]:
    df_long = df.melt(id_vars=["File"], var_name="Algorithm", value_name="Time")
    df_long = df_long.dropna()  # Remove rows with missing values

    plt.figure(figsize=(10, 6))
    for algo in algorithms_to_test.keys():
        algo_data = df_long[df_long["Algorithm"] == algo]
        plt.plot(algo_data["File"], algo_data["Time"], marker="o", label=algo)

    plt.title(f"{category} Encryption Time Comparison")
    plt.xlabel("Files")
    plt.ylabel("Encryption Time (s)")
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
